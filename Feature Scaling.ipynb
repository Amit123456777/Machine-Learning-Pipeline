# MinMaxScalar

```python
# A way to normalize the input features/variables is the Min-Max scaler. 
# By doing so, all features will be transformed into the range [0,1] meaning that the minimum and 
# maximum value of a feature/variable is going to be 0 and 1, respectively
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Acer\Downloads\Social_Network_Ads.csv")
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df1 = df.drop(["User ID", "Gender", "Purchased"], axis=1)
df1.head()
```

```python
# Before normalization
#<------------------------>

import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(df1,kde=True, hist=False)
plt.show()
```

```python
scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(df1)
normalized_data
```

```python
normalized_data = pd.DataFrame(normalized_data)
normalized_data
```

```python
# A way to normalize the input features/variables is the Min-Max scaler. By doing so, all features will be transformed into the range [0,1]
```

```python
#  using kde
# After normalization
sns.kdeplot(normalized_data)
```
