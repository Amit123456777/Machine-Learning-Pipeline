# ordinal encoder

```python
# FEATURE COLUMNS MEANS INPUT VARIABLES.
# ordinal en will apply on features
# PREDICTED COLUMN IS TARGET VARIABLE ELSE EVERYTHING IS CALLED FEATURES
```

```python
import pandas as pd
df1 = pd.DataFrame({"experience":["beginer", "advanced", "intermediate", "beginer", "advanced"],
                  "satisfaction_level": ["low", "high","medium", "medium", "low"],
                  "hrs_work": [40,50,45,30,60],
                  "performance": ["avg", "good", "good", "poor", "avg"]})
df1
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['beginer', 'intermediate', 'advanced'], ['low','medium','high']])
ordinal_encoder

# 'Beginer', 'Intermediate', 'Advanced' => 0,1,2 
```

```python
# ordinal_encoder
df1[["experience","satisfaction_level"]] = ordinal_encoder.fit_transform(df1[["experience","satisfaction_level"]])                                                               
```
